### Lidando com dados ausentes

```python
#Primeiro lidando com dados ausentes
#Ver o shape antes
shape_antes = dados.shape
print("Antes:\t", shape_antes)

dados.dropna(subset=['percentualProtestos'], inplace=True)
dados.dropna(subset=['periodoDemonstrativoEmMeses'], inplace=True)
dados.dropna(subset=['custos'], inplace=True)
dados.dropna(subset=['anoFundacao'], inplace=True)
dados.dropna(subset=['intervaloFundacao'], inplace=True)
dados.dropna(subset=['capitalSocial'], inplace=True)
dados.dropna(subset=['restricoes'], inplace=True)
dados.dropna(subset=['empresa_MeEppMei'], inplace=True)
dados.dropna(subset=['limiteEmpresaAnaliseCredito'], inplace=True)
dados.dropna(subset=['dataAprovadoNivelAnalista'], inplace=True)

#Ver o shape depois
shape_depois = dados.shape
print("Depois:\t", shape_depois)
```

```python
# Removendo duplicatas
dados.drop_duplicates(inplace=True)
dados.shape
```

### Convertendo string para float

```python
# Convertendo a coluna para float, porém nesse método cria várias colunas
aux = pd.get_dummies(dados["cnpjSemTraco"], drop_first=True)
df_final = pd.concat([dados, aux], axis=1)
```

```python
# Importação do LabelEncoder
from sklearn.preprocessing import LabelEncoder   
```

```python
# Criando o LabelEncoder
le = LabelEncoder()

# Convertendo strings em números
dados['cnpjSemTraco'] = le.fit_transform(dados['cnpjSemTraco'])
dados['nomeFantasia'] = le.fit_transform(dados['nomeFantasia'])
```
